```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Total people vs Survived

```python
titanic_df = pd.read_csv("dataset/Titanic-Dataset.csv")


# Total number of people on board
total_people = len(titanic_df)

# Total number of survivors
total_survived = titanic_df[titanic_df['Survived'] == 1].shape[0]
total_survived

# Data for visualization
data = {'Category': ['Total People', 'Survived'], 'Count': [total_people, total_survived]}

# Create a DataFrame for the data
df_visualization = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(6, 4))
sns.barplot(x='Category', y='Count', data=df_visualization, palette='viridis')
plt.title('Total People vs Survived')
plt.ylabel('Count')
plt.show()
```

##### Male vs Female in survived

```python
male_survived = titanic_df[(titanic_df['Survived'] == 1) & (titanic_df['Sex'] == "male")].shape[0]
female_survived = titanic_df[(titanic_df['Survived'] == 1) & (titanic_df['Sex'] == "female")].shape[0]


# Data for visualization
data = {'Category': ['Male Survived', 'Female Survived'], 'Count': [male_survived, female_survived]}

# Create a DataFrame for the data
df_male_female = pd.DataFrame(data)

plt.figure(figsize=(6,4))
sns.barplot(x='Category', y='Count', data=df_male_female, palette='viridis')
plt.title('Male Survived vs Female Survived')
plt.ylabel('Count')
plt.show()
```
